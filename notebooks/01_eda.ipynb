```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns

train = pd.read_csv('../data/raw/train.csv', parse_dates=['Date'])
store = pd.read_csv('../data/raw/store.csv')
df = train.merge(store, on='Store', how='left')

# Filter: open days with real sales only
df = df[(df['Open'] == 1) & (df['Sales'] > 0)].copy()
df = df.drop(columns=['Customers'])  # data leak — drop it

print(f'Shape: {df.shape}')
print(f'Date range: {df.Date.min()} to {df.Date.max()}')
print(f'Unique stores: {df.Store.nunique()}')
print(f'Nulls: {df.isnull().sum().sum()}')
df['Sales'].describe()
```

```python
daily = df.groupby('Date')['Sales'].mean().reset_index()
fig = px.line(daily, x='Date', y='Sales',
              title='Average daily sales across all stores')
fig.show()
```
